```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

model_id = "tiiuae/falcon-rw-1b"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id)

generator = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
def ask_bot(question):
    prompt = f"Answer the following general health question in simple words:\n\nQuestion: {question}\nAnswer:"
    response = generator(prompt, max_new_tokens=150, do_sample=True, temperature=0.7)[0]['generated_text']

    # Extract only the answer
    return response.split("Answer:")[-1].strip()
```

```python
while True:
    user_input = input("Ask a health question (or type 'exit'): ")
    if user_input.lower() == "exit":
        print("Goodbye!")
        break
    answer = ask_bot(user_input)
    print("Bot:", answer)
```
