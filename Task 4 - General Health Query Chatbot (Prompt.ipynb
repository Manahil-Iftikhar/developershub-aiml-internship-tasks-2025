```python
# ✅ Install required libraries
!pip install -q transformers accelerate

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

# ✅ Use small, fast model that works in Colab
model_id = "google/flan-t5-large"

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForSeq2SeqLM.from_pretrained(model_id)

# Create chatbot pipeline
chatbot = pipeline("text2text-generation", model=model, tokenizer=tokenizer)

# Prompt engineering
def build_prompt(user_input):
    return f"""You are a helpful and friendly medical assistant. Provide safe, general health-related information.
Avoid giving exact dosages or medical prescriptions. Recommend seeing a doctor for serious issues.

Question: {user_input}
Answer:"""

# Basic safety filter
def is_safe(query):
    unsafe = ["suicide", "kill", "overdose", "harm", "abortion"]
    return not any(word in query.lower() for word in unsafe)

# Chat loop
while True:
    user_input = input("Ask a health question (or type 'exit'): ")
    if user_input.lower() == 'exit':
        print("Goodbye! Stay healthy.")
        break

    if not is_safe(user_input):
        print("⚠️ Sorry, I can't answer this. Please consult a healthcare professional.")
        continue

    prompt = build_prompt(user_input)
    result = chatbot(prompt, max_new_tokens=200)
    print("\n🤖", result[0]['generated_text'].strip())
```
